```python
import torch
import cv2
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt
```

```python
model = YOLO('yolov8n-seg.yaml') 
model = YOLO('yolov8n-seg.pt')  
model = YOLO('yolov8n-seg.yaml').load('yolov8n-seg.pt')

results = model.train(data='coco128-seg.yaml', epochs=100, imgsz=640)
```

```python
model=YOLO("runs/segment/train/weights/best.pt")
```

```python
model.predict("video.mp4",save=True)
```

```python
video_path = 'video.mp4'
cap = cv2.VideoCapture(video_path)

delay = 1

detection = {}

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('video_prediction.mp4', fourcc, cap.get(5), (int(cap.get(3)), int(cap.get(4))))

while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()
        detection.clear() 

        for r in results:
            for c in r.boxes.cls.unique():
                n = (r.boxes.cls == c).sum().item() 
                detection[f"{r.names[int(c)]}"] = n

        info_text = " ".join([f"{key}: {value}" for key, value in detection.items()])
        cv2.putText(annotated_frame, info_text, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)

        out.write(annotated_frame)

        cv2.imshow("YOLOv8 Inference", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(delay) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached
        break

# Release the video capture object, close the display window, and release the video writer
cap.release()
out.release()
cv2.destroyAllWindows()
```
